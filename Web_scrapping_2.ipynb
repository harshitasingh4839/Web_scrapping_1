```python
URL = "https://books.toscrape.com/catalogue/page-1.html"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
    }
response = requests.get(URL, headers=headers)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content, 'lxml')
print(soup.prettify()) 
```

```python
book_name = soup.find_all('h3')
len(book_name)
```

```python
names = []
for i in range(len(book_name)):
    names.append(soup.find_all('h3')[i].text)
```

```python
price = soup.find_all('p',class_='price_color')
```

```python
price_list=[]
for i in range(len(price)):
    price_list.append(soup.find_all(class_='price_color')[i].text)
```

```python
avl = soup.find_all(class_='instock')
```

```python
avl_lst=[]
for i in range(len(avl)):
    avl_lst.append(soup.find_all(class_='instock')[i].text.strip())
```

```python
d = {'name':names ,
     'price':price_list,
     'availabilty':avl_lst}
```

```python
final = pd.DataFrame()

for j in range(1,30):

    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
    }
    URL = "https://books.toscrape.com/catalogue/page-{}.html".format(j)
    response = requests.get(URL, headers=headers)
    soup = BeautifulSoup(response.content, 'lxml')

    names = []
    avl_lst=[]
    price_list=[]

    book_name = soup.find_all('h3')
    
    for i in range(len(book_name)):
        names.append(soup.find_all('h3')[i].text)
        
    price = soup.find_all('p',class_='price_color')

    for i in range(len(price)):
        price_list.append(soup.find_all(class_='price_color')[i].text)
    avl = soup.find_all(class_='instock')
    
    for i in range(len(avl)):
        avl_lst.append(soup.find_all(class_='instock')[i].text.strip())
    
    df = pd.DataFrame({'name':names ,
     'price':price_list,
     'availabilty':avl_lst})

    final = pd.concat([final,df],ignore_index=True)
    


```
