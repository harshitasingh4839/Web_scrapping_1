```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
import numpy as np
```

```python
URL="https://www.tutorialsfreak.com/tutorials"
# headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"} 
webpage = requests.get(URL)

```

```python
soup = BeautifulSoup(webpage.content, 'lxml')
print(soup.prettify()) 
```

```python
desc = soup.find_all('div',class_="col-lg-4 col-sm-6")
```

```python
course_desc = []
for d in desc:
    try:
        course_desc.append(d.text)
    except:
        course_desc.append(np.nan)
```

```python
course_desc
```

```python
course = soup.find_all('a',class_='fs-20 lh-30 fw-500 label-color-5 mb-3')
course
```

```python
course_name =[]
for c in course:
    try:
        course_name.append(c.text)
    except:
        course_name.append(np.nan)
course_name
```

```python
df = pd.DataFrame({
    'course_name' : course_name,
    'desc' : course_desc
})

```
